```python
import pandas as pd
import numpy as np
```

```python
import urllib.request
webUrl  = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
data = webUrl.read()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
```

```python
table=soup.find("table")
rows=table.findAll("tr")
```

```python
t_headers = []
for tr in rows:
    # Get headers of table i.e., Rank, Country, GDP.

    for th in tr.find_all("th"):
        # remove any newlines and extra spaces from left and right
              
        t_headers.append(th.text.replace('\n', ' ').strip())

t_headers
```

```python
t_rows = []
for tr in rows:
    # Get rows of table 
    t_intermediate=[]
    for td in tr.find_all("td"):
        # remove any newlines and extra spaces from left and right
                
        t_intermediate.append(td.text.replace('\n', ' ').strip())
    t_rows.append(t_intermediate)
```

```python
df = pd.DataFrame(t_rows, columns = t_headers) 
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(df[df['Borough']=='Not assigned'].index,inplace=True)
```
